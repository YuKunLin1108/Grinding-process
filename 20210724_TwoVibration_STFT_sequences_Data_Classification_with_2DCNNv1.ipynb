### Reading Vibration files and their labels

```python
def decode_signal(audio_binary):
    raw_data = tf.strings.split(audio_binary)[74:]
    raw_data = tf.reshape(raw_data, [-1, 5])
    raw_data_1 = raw_data[:, -2]
    raw_data_1 = tf.strings.to_number(raw_data_1, out_type=tf.dtypes.float32, name=None )
    raw_data_1 = raw_data_1 - tf.math.reduce_mean(raw_data_1)
    raw_data_2 = raw_data[:, -1]
    raw_data_2 = tf.strings.to_number(raw_data_2, out_type=tf.dtypes.float32, name=None )
    raw_data_2 = raw_data_2 - tf.math.reduce_mean(raw_data_2)
    raw_data_all = tf.concat([raw_data_1, raw_data_2],-1)  # shape=(20000,)
    raw_data_all = tf.reshape(raw_data_all, [2, -1])
    return raw_data_all
```

```python
def get_label(file_path):
    parts = tf.strings.split(file_path, os.path.sep)

  # Note: You'll use indexing here instead of tuple unpacking to enable this 
  # to work in a TensorFlow graph.
    return parts[-2]
```

```python
def get_waveform_and_label(file_path):
    label = get_label(file_path)
    audio_binary = tf.io.read_file(file_path)
    waveform = decode_signal(audio_binary)
    return waveform, label
```

### Spectrogram

```python
def get_spectrogram(waveform):

  # Concatenate audio with padding so that all audio clips will be of the 
  # same length
    waveform = tf.cast(waveform, tf.float32)
    spectrogram = tf.signal.stft(
        waveform, frame_length=512,frame_step=256, fft_length=512,
        window_fn=tf.signal.hann_window, pad_end=False, name=None)

    spectrogram = tf.abs(spectrogram)
    spectrogram = spectrogram[:, :]

    return spectrogram
```

```python
def get_spectrogram_and_label_id(audio, label):
    spectrogram_1 = get_spectrogram(audio[0])
    spectrogram_1 = tf.expand_dims(spectrogram_1, -1)
    spectrogram_2 = get_spectrogram(audio[1])
    spectrogram_2 = tf.expand_dims(spectrogram_2, -1)
#     spectrogram = tf.data.Dataset.zip((spectrogram_1, spectrogram_2))
#     spectrogram = tf.concat([spectrogram_1, spectrogram_2], -1)
    label_id = tf.argmax(label == commands)
    return (spectrogram_1, spectrogram_2), label_id
```

### Build and train the model

```python
def preprocess_dataset(files):
    files_ds = tf.data.Dataset.from_tensor_slices(files)
    output_ds = files_ds.map(get_waveform_and_label, num_parallel_calls=AUTOTUNE)
    output_ds = output_ds.map(
        get_spectrogram_and_label_id,  num_parallel_calls=AUTOTUNE)
    return output_ds
```

```python
train_ds = spectrogram_ds
val_ds = preprocess_dataset(val_files)
test_ds = preprocess_dataset(test_files)
```

```python
batch_size = 8
train_ds = train_ds.batch(batch_size)
val_ds = val_ds.batch(batch_size)
```

```python
train_ds = train_ds.cache().prefetch(AUTOTUNE)
val_ds = val_ds.cache().prefetch(AUTOTUNE)
```

```python
for (spectrogram1, spectrogram2), _ in spectrogram_ds.take(1):
    input_shape1 = spectrogram1.shape
    input_shape2 = spectrogram2.shape
print('Input 1 shape:', input_shape1)
print('Input 2 shape:', input_shape2)
num_labels = len(commands)
```

```python
# norm_layer = preprocessing.Normalization()
# norm_layer.adapt(spectrogram_ds.map(lambda x, _: x))
```

```python
from tensorflow import keras
# Input Layer
v1_input = keras.Input(shape=input_shape1, name='V1_Input')
v2_input = keras.Input(shape=input_shape2, name='V2_Input')

# Hidden Layer
h1_1 = layers.Conv2D(32,3,strides=1,activation='relu', name='hidden1_1')(v1_input)
h1_2 = layers.Conv2D(32,3,strides=1,activation='relu', name='hidden1_2')(h1_1)
h1_2_ft = layers.Flatten()(h1_2)

h2_1 = layers.Conv2D(32,3,strides=1,activation='relu', name='hidden2_1')(v2_input)
h2_2 = layers.Conv2D(32,3,strides=1,activation='relu', name='hidden2_2')(h2_1)
h2_2_ft = layers.Flatten()(h2_2)

concat = layers.Concatenate()([h1_2_ft, h2_2_ft])
h3 = layers.Dense(64, activation='relu', name='hidden3')(concat)

# Output Layer
outputs = layers.Dense(num_labels, name='Output')(h3)

# Create Model
model = keras.Model(inputs=[v1_input, v2_input], outputs=outputs)

model.summary()
```

```python
# from tensorflow.keras.utils import plot_model
# from IPython.display import Image
# # show structure of model
# plot_model(model, to_file='Function_API_Multi_Input_Model.png')
# Image('Function_API_Multi_Input_Model.png')
```

```python
from tensorflow.keras import callbacks
log_dirs = 'lab1-log-scalers'
model_cbk = callbacks.TensorBoard(log_dirs)
```

```python
model.compile(
    optimizer=tf.keras.optimizers.Adam(),
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics=['accuracy'],
)
```

```python
EPOCHS = 10
history = model.fit(
    train_ds, 
    validation_data=val_ds,  
    epochs=EPOCHS,
#     callbacks=tf.keras.callbacks.EarlyStopping(verbose=1, patience=2),
    callbacks=[tf.keras.callbacks.EarlyStopping(verbose=1, patience=2), model_cbk],
)
```

```python
metrics = history.history
plt.plot(history.epoch, metrics['loss'], metrics['val_loss'])
plt.legend(['loss', 'val_loss'])
plt.show()
```

### Evaluate test set performance

```python
test_audio1 = []
test_audio2 = []
test_labels = []

for (audio1, audio2), label in test_ds:
    test_audio1.append(audio1.numpy())
    test_audio2.append(audio2.numpy())
    test_labels.append(label.numpy())
    
test_audio1 = np.array(test_audio1)
test_audio2 = np.array(test_audio2)
test_labels = np.array(test_labels)
```

```python
y_pred = np.argmax(model.predict((test_audio1, test_audio2)), axis=1)
y_true = test_labels

test_acc = sum(y_pred == y_true) / len(y_true)
print(f'Test set accuracy: {test_acc:.0%}')
```
